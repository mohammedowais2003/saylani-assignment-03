```python
#1) Write the code to print the following pattern.

n = 3

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")

    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()


```

```python
#2) Take user input and write the code to print all the prime numbers till the user input.

num = int(input("Enter a number: "))

for i in range(2, num + 1):
    count = 0

    for j in range(1, i + 1):
        if i % j == 0:
            count += 1

    if count == 2:
        print(i)
```
